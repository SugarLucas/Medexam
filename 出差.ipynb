```python
import os
from openpyxl import Workbook

folder_path = '/Users/lucas007/Desktop/intern/出院'
output_path = '/Users/lucas007/Desktop/intern/出院诊断.xlsx'

# Create a new workbook and sheet
workbook = Workbook()
sheet = workbook.active
sheet.title = 'Diagnoses'

# Write the column header
sheet.cell(row=1, column=1, value='姓名')
sheet.cell(row=1, column=2, value='入院日期')
sheet.cell(row=1, column=3, value='出院日期')
sheet.cell(row=1, column=4, value='PATIENT_ID')
sheet.cell(row=1, column=5, value='入院诊断')
sheet.cell(row=1, column=6, value='出院诊断')
sheet.cell(row=1, column=7, value='出院医嘱')
sheet.cell(row=1, column=8, value='年龄')
sheet.cell(row=1, column=9, value='性别')


# Initialize the current row to 2 (since the first row is the header)
current_row = 2
for root, dirs, files in os.walk(folder_path):
    for file in files:
        file_path = os.path.join(root, file)
        file_name = os.path.basename(file_path)
        parts = file_name.split('-')
        if len(parts) >= 2:
            code = parts[1]
            sheet.cell(row=current_row, column=4, value=code)
        try:
            with open(file_path, "r", encoding='gbk', errors='replace') as file:
                for line in file:  
                    if "姓名" in line:
                        #name = line.split("：")[1].strip()
                        # Write the diagnosis to the sheet
                        # Find the position of the colon character after "姓名"
                        start =line.find('姓名：') + len('姓名：')
                        end = line.find(' ', start)

                        # Extract the substring between the colon and the next space character
                        name = line[start:end]
                        sheet.cell(row=current_row, column=1, value=name)
                    if "入院日期" in line:
                        in_date = line.split("：")[1].split()[0].strip()
                        if not in_date:
                            print(in_date)
                            next_line = next(file, '').strip()
                            while not next_line:
                                next_line = next(file, '').strip()
                            in_date = next_line
                        # Write the diagnosis to the sheet
                        sheet.cell(row=current_row, column=2, value=str(in_date))
                    if "出院日期" in line:
                        out_date = line.split("：")[1].split()[0].strip()
                        if not out_date:
                            print(out_date)
                            next_line = next(file, '').strip()
                            while not next_line:
                                next_line = next(file, '').strip()
                            out_date = next_line
                        # Write the diagnosis to the sheet
                        sheet.cell(row=current_row, column=3, value=str(out_date))
                    if "入院诊断" in line:
                        in_diagnoses = line.split("：")[1].strip()
                        # Write the diagnosis to the sheet
                        if not in_diagnoses:
                            next_line = next(file, '').strip()
                            if next_line:
                                in_diagnoses = next_line
                        sheet.cell(row=current_row, column=5, value=in_diagnoses)
                    if "出院诊断" in line:
                        diagnoses = line.split("：")[1].strip()
                        if not diagnoses:
                            next_line = next(file, '').strip()
                            if next_line:
                                diagnoses = next_line
                        # Write the diagnosis to the sheet
                        sheet.cell(row=current_row, column=6, value=diagnoses)
                    if "出院医嘱" in line:
                        instruction = line.split("：")[1].strip()
                        # Write the diagnosis to the sheet
                        sheet.cell(row=current_row, column=7, value=instruction)
                    if "年龄" in line:
                        #name = line.split("：")[1].strip()
                        # Write the diagnosis to the sheet
                        # Find the position of the colon character after "姓名"
                        start =line.find('年龄：') + len('年龄：')
                        end = line.find(' ', start)

                        # Extract the substring between the colon and the next space character
                        name = line[start:end]
                        sheet.cell(row=current_row, column=8, value=name)
                    if "性别" in line:
                        #name = line.split("：")[1].strip()
                        # Write the diagnosis to the sheet
                        # Find the position of the colon character after "姓名"
                        start =line.find('性别：') + len('性别：')
                        end = line.find(' ', start)

                        # Extract the substring between the colon and the next space character
                        name = line[start:end]
                        sheet.cell(row=current_row, column=9, value=name)
                        current_row += 1
        except Exception as e:
            print(f"Error processing file {file_path}: {e}")
            continue
# Save the workbook to a file
workbook.save(output_path)
```

```python
# Split the column values and do a frequency count
diagnosis_counts = {}
for value in final["出院诊断"].dropna():
    # Split the value by numbers and dots using a regular expression
    diagnoses = re.split("[0-9]+\\.", value)
    # Remove empty strings and whitespace from the diagnoses list
    diagnoses = [diag.strip() for diag in diagnoses if diag.strip()]
    # Count the frequency of each diagnosis
    for diag in diagnoses:
        diagnosis_counts[diag] = diagnosis_counts.get(diag, 0) + 1


# Create a new DataFrame to store the diagnosis counts
diagnosis_df = pd.DataFrame({
    "诊断": list(diagnosis_counts.keys()),
    "频数": list(diagnosis_counts.values())
})

# Sort the diagnosis counts by count in descending order
diagnosis_df = diagnosis_df.sort_values("频数", ascending=False)

# Print the diagnosis counts
print(diagnosis_df)
```
